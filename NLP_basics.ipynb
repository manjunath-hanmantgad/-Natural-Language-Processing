```python
# import spacy and start using ready-to-go core models in spacy

import spacy
nlp = spacy.load('en_core_web_trf')
```

 Perform Tokenization

Tokenization will segment text into words, punctuation marks, etc.



```python
# invoke the tokenizer using nlp.tokenizer(sentence)

sentence = nlp.tokenizer("A walk to remmember is a good movie to watch.")

# length of sentence

print("The number of tokens:", len(sentence))

# print individual words 

print("The tokens : ")
for words in sentence:
  print(words)
```

Notice above full stop (.) is also a token.

Loading publicly available datasets for more practice.

```python
import pandas as pd
import os
cwd = os.getcwd()

# import jeopardy questions from jeopardy dataset

data = pd.read_csv(cwd+'/data/jeopardy_questions/jeopardy_questions.csv')
data = pd.DataFrame(data=data)

# lowercase , remove/strip whitespace and view the columns

data.columns = map(lambda x: x.lower().strip() , data.columns)

# reduce the size of data

data = data[0:1000]

# tokenize the jeopardy questions 

data["question_tokens"] = data["question"].apply(lambda x: nlp(x))
```

**Chunking**

By grouping related tokens into chunks, the machine will have an easier time processing the sentence. Instead of viewing each token in isolation, the machine now recognizes that certain tokens are related to others,

```python
for token in nlp("The night is darker at night."):
  print(token.text)
```

```python
# print chunks now

for chunk in nlp("The night is darker at night.").noun_chunks:
  print(chunk.text)
```
