```python
# using displacy

import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_md')

doc = nlp('I like doing this.')

displacy.serve(doc , style='dep')
```

```python
# Entity recognizer 
#import spacy
#from spacy import displacy
nlp = spacy.load('en_core_web_md')
doc = nlp("Tim Cook is Apple's CEO in 2021.")
displacy.render(doc, style='ent')# using .render inplace of .serve as Im already using a browser.
```

```python
# to save a file outside for presentations or sharing 

import spacy
from spacy import displacy

from pathlib import Path
nlp = spacy.load('en_core_web_md')
doc = nlp('This is a bird.')
svg = displacy.render(doc , style='dep' , jupyter=False)
filename = 'bird.svg'

output_path = Path('/images' + filename)
output_path.open('w' , encoding='utf-8').write(svg)
```
